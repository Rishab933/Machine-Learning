```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.head()
```



*   Age - Numerical column (does not have missing values)
*   fever - Numerical column --> simple imputer to fill the  missing values
*   city & gender - Nominal categorical data --> OHE
*   review & cough - Ordinal categorical data --> Ordinal Encoding
*   has_covid - output column --> label encoding



```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

```python
x_train.head()
```

```python
from sklearn.compose import ColumnTransformer
```

remainder can be


1.   passthrough --> skip or leae that column as it is.
2.   drop --> drop that column, that means after encoding that column won't be present in the output



transformer = ColumnTransformer(transformers=['trans_name', 'encoding_type()',['column_name']],remainder='passthrough/drop')

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(),['fever']),
    ('tnf2', OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3', OneHotEncoder(sparse=False, drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(x_train)
```

```python
transformer.fit_transform(x_train).shape
```

```python
transformer.transform(x_test)
```

```python
transformer.transform(x_test).shape
```
