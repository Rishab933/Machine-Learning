PCA aims to scale down the higher dimensional data to a lower dimensional data keeping the essance of the data.

It takes the smallest possible dimension.

It results in faster execution of algo and it helps to visualize a higher dimensional data as it converts the whole data in a lower dimensional as 3D or 2D, because these two can only be observed by humans not more than this.

For a non known data set if we have to select some features of the data for our model then plot scatter plot of two columns and see data spread is more for which feature.

eg: we have three columns as **No. of Room**, **No. of Grocery shop**, **Price** i have to select some of the features out of these three.

for this i will plot scatter plot for **room** and **grocery** and will observe where the data spread is more.

the feature the most data spread is taken and the other is removed.

If the data set has **No of Room**, **No of washrooms**, **Price** then here we cant do feature selection as spread of both the feature on both the axis is almost comparable to each other

Therefore we will apply feature extraction --> Creating a new feature as **size of the flat** now here this feature is equivalent to **no of room** and **no of washroom**

Variance is used for PCA.

Number of input column is = num of dimension.

**step 1:** mean centering with respect to every axis(dimension).

**step 2:** Find the co variance matrix
The matrix will be dimension x dimension
**eg:** we have f1, f2, f3 as the input columns then.

v(f1)      c(f2.f1)   c(f1.f3)

c(f1.f2)   v(f2)      c(f2.f3)

c(f1.f3)   c(f2.f3)   v(f3)

**step 3:** Find the eigen value and eigen vector for the co variance matrix.

For the above example we will get a set of 3 eigen values and a set of 3 eigen vectors say X1, X2, X3 and in this set of 3 vectors the vector which has the highest eigen value that is refered as PC1 then PC2 and then PC3.

Now if we consider only PC1 then dimension is reduced to 1D, if PC1 and PC2 is considered then dimension is reduced to 2D.

```python
import numpy as np
import pandas as pd

np.random.seed(23)

mu_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20)

df = pd.DataFrame(class1_sample,columns=['feature1','feature2','feature3'])
df['target'] = 1

mu_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2_sample = np.random.multivariate_normal(mu_vec2, cov_mat2, 20)

df1 = pd.DataFrame(class2_sample,columns=['feature1','feature2','feature3'])

df1['target'] = 0

df = pd.concat([df, df1], ignore_index=True)
df = df.sample(40)
```

```python
df.head()
```

```python
import plotly.express as px
#y_train_trf = y_train.astype(str)
fig = px.scatter_3d(df, x=df['feature1'], y=df['feature2'], z=df['feature3'],
              color=df['target'].astype('str'))
fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))

fig.show()
```

```python
# Step 1 - Apply standard scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df.iloc[:,0:3] = scaler.fit_transform(df.iloc[:,0:3])
```

```python
# Step 2 - Find Covariance Matrix
covariance_matrix = np.cov([df.iloc[:,0],df.iloc[:,1],df.iloc[:,2]])
print('Covariance Matrix:\n', covariance_matrix)
```

```python
# Step 3 - Finding EV and EVs
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
pc = eigen_vectors[0:2]
pc
```

```python
transformed_df = np.dot(df.iloc[:,0:3],pc.T)
# 40,3 - 3,2
new_df = pd.DataFrame(transformed_df,columns=['PC1','PC2'])
new_df['target'] = df['target'].values
new_df.head()
```

```python
new_df['target'] = new_df['target'].astype('str')
fig = px.scatter(x=new_df['PC1'],
                 y=new_df['PC2'],
                 color=new_df['target'],
                 color_discrete_sequence=px.colors.qualitative.G10
                )

fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig.show()
```
