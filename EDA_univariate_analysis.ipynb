# **Countplot**

```python
import seaborn as sns
```

```python
print(df['Survived'].unique())
```

```python
import matplotlib.pyplot as plt

sns.countplot(x='Survived', data=df)

plt.xlabel('Survived')  # Set x-axis label
plt.ylabel('Count')     # Set y-axis label
plt.title('Survival Count on the Titanic')  # Set plot title
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'])  # Customize x-ticks

plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.xlabel('Passenger class')
plt.ylabel('Count')
plt.title('Passenger class Distribution')
plt.xticks(ticks=[0, 1, 2], labels=['First class', 'Second class', 'Third class'])
plt.show()
```

```python
sns.countplot(x='Embarked', data=df, palette=['red', 'green', 'blue'])
plt.xlabel('Embarked')
plt.ylabel('count')
plt.title('Embarked Distribution')
plt.xticks(ticks=[0,1,2], label=['S','C','Q'])
plt.show()
```

df['Survived'].value_counts().plot(kind='bar') --> To generate a bar char

# **Pie chart**

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
plt.title('Survival count on titanic')
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
plt.title('Passenger class Distribution')
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%.2f')
plt.title('Sex Distribution')
plt.show()
```

```python
df['Embarked'].value_counts().plot(kind='pie', autopct='%.2f')
plt.title('Embarked Distribution')
```

# **Histogram**

```python
plt.hist(df['Age'], bins=50)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# **Distplot**
KDE = kernal density estimation

PDF = probability density function

```python
sns.distplot(df['Age'])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Prob')
plt.show()
```

# **Boxplot**
It is used to eliminate noisy data

```python
sns.boxplot(x = df['Fare'])
plt.title('Fare Distribution')
```

```python
sns.boxplot(x = df['Age'])
plt.title('Fare Distribution')
```

The point that are to the right are called outliers its different than the normal therefore they can be eliminated.
