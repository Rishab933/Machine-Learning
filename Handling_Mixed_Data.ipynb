```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
# here A is for alone that means travelling alone
```

```python
fig = df['number'].value_counts().plot(kind='bar')
fig.set_title('Passenger travelling with')
```

# **Number column**

```python
# extracting numerical part from the column number
df['number_numerical'] = pd.to_numeric(df["number"],errors='coerce',downcast='integer')
```

```python
# extracting categorical part from the column number
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)
df.head()
```

```python
df['number_numerical'].fillna(0,inplace=True)
df['number_categorical'].fillna('Not_alone',inplace=True)
```

```python
df.head()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') #this ('(\d+)') is used to extract number from a string
df['cabin_cat'] = df['Cabin'].str[0]
df.head()
```
