# **Date in both the dataframes is an object data type so first we have to convert it in date time variable**

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

# **Extracting year from the date**

```python
date['date_year'] = date['date'].dt.year
```

```python
date.head(4)
```

# **Extracting Month**

```python
date['date_month'] = date['date'].dt.month_name()
date.sample(4)
```

# **Extracting days**

```python
date['date_day'] = date['date'].dt.day
date['day_name'] = date['date'].dt.day_name()
date.sample(4)
```

```python
#Checking if the day is a weekend or not?
# .isin([this_list_then]),first_arg, or_else_second_arg
date['date_is_weekend'] = np.where(date['day_name'].isin(['Sunday', 'Saturday']), 1,0)
date.drop(columns=['product_id', 'city_id', 'orders']).head(4)
```

# **Extracting week number of the year**

```python
date['week_number'] = date['date'].dt.isocalendar().week
date.drop(columns=['product_id', 'city_id', 'orders']).head(4)
```

# **Extracting quarter**

```python
date['quater'] = date['date'].dt.quarter
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

# **Extracting semesters**

```python
date['semester'] = np.where(date['quater'].isin([1,2]),1,2)
date.drop(columns=['product_id','city_id','orders']).head()
```

# **Finding gap between two dates**

```python
import datetime
```

```python
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
#want to have only the umber of days
(today - date['date']).dt.days
```

```python
months_passed = (today - date['date']) / np.timedelta64(1, 'D')
months_passed = np.round(months_passed / 30, 0).astype(int)
months_passed
```
